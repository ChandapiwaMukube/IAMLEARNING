```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
image = imread('SCOOPY.png')
```

```python
image.shape
```

```python
plt.imshow(image)
plt.title('SCOOPY IMAGE')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```

```python
x = image.reshape(-1,3)
image.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state =45)

kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
seg_image = kmeans.cluster_centers_[kmeans.labels_]
```

```python
seg_image = seg_image.reshape(image.shape)
seg_image.shape
```

```python
plt.imshow(seg_image)
plt.title('Segmented image ')
plt.xlabel('Width')
plt.ylabel('Length')
plt.show()
```
